```python
from bs4 import BeautifulSoup
from curl_cffi import requests as req
import pandas as pd
import csv
```

```python
header = {"User_Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:124.0) Gecko/20100101 Firefox/124.0"}
```

```python
html_text = req.get("https://apna.co//jobs/jobs-in-mumbai", impersonate='chrome').text
```

```python
soup = BeautifulSoup(html_text, 'lxml')
jobs = soup.find_all('div', class_='mx-[12px] md:m-0')
c = 0
records = []
for i in jobs:
    t = []
    c+=1
    t.append(c)
    header = i.find('h2').text
    t.append(header)
    
    company = i.find(class_='JobListCardstyles__JobCompany-ffng7u-8 gguURM').text 
    t.append(company)
    
    loc = i.find(class_='m-0 text-sm leading-[20px] text-[#8C8594]').text 
    t.append(loc)
    
    salary = i.find(class_='m-0 truncate text-sm leading-[20px] text-[#8C8594]').text
    t.append(salary)
    require = []
    for j in i.find_all('p', class_='m-0 whitespace-nowrap text-xs text-[#8C8594]'):
        require.append(j.text)
    t.append(require)

    records.append(t)

    
with open('jobs.csv', 'a', newline='\n') as csvf:
    upload = csv.writer(csvf, delimiter=',')
    upload.writerows(records)
```
